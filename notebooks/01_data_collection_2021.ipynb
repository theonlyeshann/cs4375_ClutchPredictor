# Data Collection: NBA Season 2021 - 2022:
- Winners: Golden State Warriors

```python
from nba_api.stats.endpoints import shotchartdetail
from nba_api.stats.static import players
import pandas as pd
import time
```

```python
def get_player_id(player_name):
    all_players = players.get_players()
    for p in all_players:
        if p['full_name'] == player_name:
            return p['id']
    return None
```

```python
def get_shot_data(player_id, season='2021-22'):
    response = shotchartdetail.ShotChartDetail(
        team_id=0,
        player_id=player_id,
        season_nullable=season,
        context_measure_simple='FGA'
    )
    return response.get_data_frames()[0]
```

```python
def get_multiple_players_data(player_names, season='2021-22'):
    all_data = []

    for name in player_names:
        try:
            player_id = get_player_id(name)
            df = get_shot_data(player_id, season)
            df['PLAYER_NAME'] = name
            all_data.append(df)
            time.sleep(1)
        except Exception as e:
            print(f"Error with {name}: {e}")

    return pd.concat(all_data, ignore_index=True)
```

```python
players_list = [
    "LeBron James",
    "Stephen Curry",
    "Kevin Durant",
    "Luka Dončić",
    "Jayson Tatum",
    "Ja Morant",
    "Nikola Jokić",
    "Giannis Antetokounmpo",
    "Joel Embiid",
    "Anthony Davis",
    "Kawhi Leonard",
    "Damian Lillard",
    "Devin Booker",
    "Zion Williamson",
    "Trae Young",
    "Anthony Edwards",
    "Shai Gilgeous-Alexander",
    "Karl-Anthony Towns",
    "Klay Thompson",
    "Jimmy Butler",
    "Aaron Gordon",
    "Victor Wembanyama",
    "Jaylen Brown",
    "De'Aaron Fox",
    "Brandon Ingram",
    "Pascal Siakam",
    "Bam Adebayo",
    "Rudy Gobert",
    "Donovan Mitchell",
    "Otto Porter Jr.",
    "Jordan Poole",
]

df = get_multiple_players_data(players_list)
df.head()
```
